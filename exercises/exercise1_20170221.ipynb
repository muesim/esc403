# Short python demo

First we import numpy for numerical analysis.

```python
#   To execute a cell:   type  "ctrl-enter"  (or "shift-enter" to also select the next cell)
import numpy as np
```

```python
# create a numpy variable and perform a vector operation
x = np.arange(0., 20, .1)
```

```python
# Perform a vector operation
kappa = 0.2
y = np.exp(-kappa*x)*np.sin(x)
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y, color='purple', label='dampened oscillator', lw=3)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```
